### Numerical Analysis of Algorithms for a Small Function
##$f(x)=0.1X_1^2+2X_2^2$

We will plot the Level sets of this funtions. Then Numerically and Visaully analyse different optimization alorithms for finding the optimal solution.

Later on we will extend this idea to Neural Networks for Image Classification Task.

##$âˆ‡f(x)=[0.2X_1,4X_2]^T$


##$v_{t}=\rho  v_{t-1}+ g_t$

##$w_{new}=w_{old}- \alpha v_{t}$


##$w_{2}=w_{1}-\alpha g_{1}$

##$w_{3}=w_{2}-\alpha ( \rho g_{1}+g_{2})$

##$w _{4}=w_{3}-\alpha ( \rho ^{2}g_{1}+ \rho g_{2}+g_{3})$

```
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(13,6))
ax = plt.axes(projection="3d") 
start, stop, n_values = -8, 8, 800
x_vals = np.linspace(start, stop, n_values)
y_vals = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x_vals, y_vals)
Z = np.sqrt(0.1*X**2 + 2*Y**2)
plt.contourf(X,Y,Z,)
#plt.colorbar(cp)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='winter', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

###RMSprop

```
c=1e-6
c
import math
```

```
v1,v2=0,0
gamma=0.9
lr=0.4
x1,x2=-4,-6
l1,l2=[],[]
for i in range(20):
  l1.append(x1)
  l2.append(x2)
  v1=gamma*v1+(1-gamma)*(0.2*x1)**2
  v2=gamma*v2+(1-gamma)*(4*x2)**2
  x1=x1-(lr/math.sqrt(v1+c))*0.2*x1
  x2=x2-(lr/math.sqrt(v2+c))*4*x2
  
  
  print(x1,x2)
```

### ADAGRAD



##$G_{t}=G_{t-1}+dw^{2}$

##$w_{new}=w_{old}-\dfrac {\alpha }{\sqrt {G_{t}+\epsilon}}\odot dw$



```
v1,v2=0,0
gamma=0.9
lr=0.4
x1,x2=-4,-6
l1_adagrad,l2_adagrad=[],[]
for i in range(20):
  l1_adagrad.append(x1)
  l2_adagrad.append(x2)
  v1=v1+(0.2*x1)**2
  v2=v2+(4*x2)**2
  x1=x1-(lr/math.sqrt(v1+c))*0.2*x1
  x2=x2-(lr/math.sqrt(v2+c))*4*x2
  
  
  
```

##$v_{t}=\rho v_{t-1}+\left( 1-\rho \right) \cdot dw^{2}$

##$w_{new}=w_{old}-\dfrac {\alpha }{\sqrt {v_{t}+\epsilon}}\cdot dw$

###Momentum

```
v1,v2=0,0
gamma,lr=0.5,0.4
x1,x2=-4,-6
l1_gd_m,l2_gd_m=[],[]
for i in range(20):
  l1_gd_m.append(x1)
  l2_gd_m.append(x2)
  v1=gamma*v1+(0.2*x1)
  v2=gamma*v2+(4*x2)
  x1=x1-lr*v1
  x2=x2-lr*v2
  
  
  
```

##$v_{t}=\rho_1 v_{t-1}+\left( 1-\rho_1 \right) dw$

##$s_{t}=\rho_2 s_{t-1}+\left( 1-\rho_2 \right) \cdot dw^{2}$

##$\hat v_t=\dfrac {v_{t}}{1-\rho_1^{t}}$

##$\hat s_t=\dfrac {s_{t}}{1-\rho_2^{t}}$

##$w_{new}=w_{old}-\dfrac {\alpha \hat v_t}{\sqrt {\hat s_t}+\epsilon}$

where $\rho_1$ and $\rho_2$ are weighting parameters; $\hat v_t$ and $\hat s_t$ are normalized state variables.

```
v1,v2,s1,s2,gamma,beta,lr=0,0,0,0,0.9,0.999,0.4
x1,x2=-6,-6
l1_adam,l2_adam=[],[]
for i in range(20):
  l1_adam.append(x1)
  l2_adam.append(x2)
  v1=gamma*v1+(1-gamma)*(0.2*x1)
  v2=gamma*v2+(1-gamma)*(4*x2)
  s1=beta*s1+(1-beta)*(0.2*x1)**2
  s2=beta*s2+(1-beta)*(4*x2)**2
  m_hat_v1= v1 / (1 - np.power(gamma, i+1))
  m_hat_v2 = v2 / (1 - np.power(gamma, i+1))
  s_hat_s1= s1 / (1 - np.power(beta, i+1))
  s_hat_s2 = s2 / (1 - np.power(beta, i+1))
  x1=x1-(lr)*(m_hat_v1/math.sqrt((s_hat_s1)+c))
  x2=x2-(lr)*(m_hat_v2/math.sqrt((s_hat_s2)+c))
  print(x1,x2)
  if abs(x1)<=0.1 and abs(x2)<0.1:
    break
  
  
```
